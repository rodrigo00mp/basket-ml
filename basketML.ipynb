# 1 Data Collection #

In this project the data will come from NBA Stats API.

```python
import requests
import pandas as pd
```

```python

api_url = os.getenv("API_URL")
teams_url = api_url + "teams"
api_key = os.getenv("API_KEY")

headers = {
	"x-rapidapi-key": f"{api_key}",
	"x-rapidapi-host": "api-nba-v1.p.rapidapi.com"
}

response = requests.get(teams_url, headers=headers)
```

```python
data = response.json()
teams = response.json()["response"]
teams_Df = pd.DataFrame(teams)
teams_Df.head()
```

```python
nbaTeams_Df = teams_Df[teams_Df['nbaFranchise']].reset_index(drop=True)
nbaTeams_Df.shape
```

```python
nbaTeams_Df["conference"] = nbaTeams_Df["leagues"].apply(lambda x: x["standard"]["conference"] if isinstance(x, dict) and "standard" in x else None)
nbaTeams_Df["division"] = nbaTeams_Df["leagues"].apply(lambda x: x["standard"]["division"] if isinstance(x, dict) and "standard" in x else None)
```

```python
nbaTeams_Df = nbaTeams_Df.drop(columns=['allStar', 'leagues', 'nickname', 'code', 'logo', 'city', 'nbaFranchise'], axis=1)

nbaTeams_Df
```

```python
nbaTeams_Df = nbaTeams_Df[nbaTeams_Df['name'] != 'Home Team Stephen A']
nbaTeams_Df.shape
```
